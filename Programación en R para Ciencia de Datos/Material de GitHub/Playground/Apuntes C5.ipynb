```r
set.seed(15)
data <- data.frame(x = rnorm(1000))
head(data)
```

```r
data %>% 
    ggplot() + 
    aes(x = x, y = after_stat(density)) +
    geom_histogram(
        # bins = 10,
        binwidth = 1,  # Con esta opción, las barras y las etiquetas sí permiten sumar 1.
        fill = 'green',
        col = 'red'
    ) +
    stat_bin(
        # bins = 10,
        binwidth = 1,  # Con esta opción, las barras y las etiquetas sí permiten sumar 1.
        geom = 'label',
        color = 'blue',
        mapping = aes(label = after_stat(round(density, 2))),
        position = position_stack(vjust = 1)
    )
```
