```r
library(ggplot2)
library(dplyr)
```

```r
# Distribución normal

# Asumiendo que la altura de los varones en Chile sigue una distribución normal con una media de 171,8 cms y
# desviación estándar de 5 cms.

# ¿Cuál es la probabilidad de encontrar un varón menor o igual a 160 cms?
pnorm(160, mean = 171.8, sd = 5)
```

```r
# ¿Cuál es la probabilidad de encontrar un varón mayor de 160 cms?
resp2a = 1 - pnorm(160, mean = 171.8, sd = 5)
resp2a

resp2b = pnorm(160, mean = 171.8, sd = 5, lower.tail = FALSE)
resp2b

# En el caso de una distribuión contínua, "<" == "<=" & ">" == ">=", por demostración...
```

```r
# ¿Cuál es la probabilidad de que un varón se encuentre entre los 180 y 190 cms de estatura?

pnorm(190, mean = 171.8, sd = 5) - pnorm(180, mean = 171.8, sd = 5)
```

```r
# ¿Cuál es la estatura que corresponde al 90% de los varones de menor tamaño?

qnorm(0.9, mean = 171.8, sd = 5)
```

```r
# Genere una muestra aleatoria de tamaño 1.000 con la media y desviación estándar entregadas anteriormente.

set.seed(2020) # Semilla
estaturas <- rnorm(1000, mean = 171.8, sd = 5)
head(estaturas)
```

```r
# Recordar que son observaciones aleatorias, por lo que la media y la desviación estándar pueden variar levemente.

mean(estaturas)
sd(estaturas)
```

```r
hist(estaturas)
```

```r
hist(estaturas, freq = FALSE)
```

```r
hist(estaturas, freq = FALSE, 30)
```

```r
dnorm(1, mean = 0, sd = 1)
```

```r
# Para mostrar la curva, se debe llamar junto con el histograma.

hist(estaturas, freq = FALSE, 30)
curve(dnorm(x, mean = 171.8, sd = 5), add = TRUE, col = '#FF0000')
curve(dnorm(x, mean = mean(estaturas), sd = sd(estaturas)), add = TRUE, col = '#3333FF')
```

```r
est_df <- data.frame(estaturas = estaturas)
head(est_df)
```

```r
# El eje y es el conteo.

est_df %>% 
    ggplot() +
    aes(x = estaturas) +
    geom_histogram(fill = 'grey', color = 'black')
```

```r
# El eje y es la densidad, de total de la muestra = 1

est_df %>% 
    ggplot() +
    aes(x = estaturas) +
    geom_histogram(
        fill = 'grey',
        color = 'black',
        aes(y = after_stat(density))
    ) +
    geom_density(colour = '#FF0000') + # Grafica una línea de densidad no paramétrica.
    geom_function(
        fun = dnorm,
        args = list(mean = 171.8, sd = 5),
        colour = '#3333FF'
    ) + # Grafica una curva de distribución normal con los datos originales.
    geom_function(
        fun = dnorm,
        args = list(mean = mean(estaturas), sd = sd(estaturas)),
        colour = '#33FF33'
    ) # Grafica una curva de distribución normal con los datos de la muestra.
    
```

```r
# Distribución binomial

# Suponga que un spot publicitario tiene una tasa de efectividad des 20% para una población determinada.
# Si se logra llegar con el spot a 100 personas ¿cuál es la probabilidad de que, al menos, 30 personas compren?

resp2a <- 1 - pbinom(29, size = 100, p = 0.2) # En caso de mayor o igual, sí es necesario "casos exitosos - 1"
resp2a

resp2b <- pbinom(29, size = 100, p = 0.2, lower.tail = FALSE)
resp2b
```

```r
# ¿Cuál es la probabilidad de que exactamente 30 personas compren?

pbinom(30, size = 100, p = 0.2, ) - pbinom(29, size = 100, p = 0.2)
```

```r
# Supongamos que queremos lanzar una moneda honesta (p(cara) = p(sello)) 14 veces.
# ¿Cuál es la probabilidad de obtener exactamente 4 caras?

dbinom(4, size = 14, p = 0.5)
```

```r
# ¿Cuál es la probalidad de obtener, al menos, 10 caras?

pbinom(9, size = 14, p = 0.5, lower.tail = FALSE)
```

```r
# Simulación de un experimento binomial.
# Si en 10.000 experimentos lanzamos una moneda honesta 14 veces, se anota el número de veces de éxito,
# por ejemplo, obtener caras.

set.seed(2019)
caras <- rbinom(10000, size = 14, p = 0.5)
caras
```

```r
mean(caras) / 14
```

```r
# ¿En qué porcentaje de los lanzamientos se obtuvo 4 caras?

caras_4 <- mean(caras == 4) * 100
paste(caras_4, '%')
```

```r
monedas_df <- data.frame(caras = caras)

monedas_df %>% 
    ggplot() + 
    aes(x = caras) +
    geom_bar()
```

```r
N = 10000 # Ojo, revisar de dónde sale 'N' realmente, porque el gráfico sale maluenda.

monedas_df %>%
    count(caras) %>% 
    bind_rows(data.frame(caras = 14, n = 0)) %>% # En el caso de esta muestra no había casos de 14 caras por lo que se agrega
    mutate(
        freq_relat_obs = n / N,
        prob_real = dbinom(caras, 14, 0.5)
    ) %>% 
    ggplot() +
    aes(x = caras, y = freq_relat_obs) +
    geom_bar(stat = 'identity') +
    geom_point(
        aes(y = prob_real),
        colour = '#3333FF',
        size = 4
    )
```
