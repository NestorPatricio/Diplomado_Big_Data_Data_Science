```r
df <- data.frame(
    Edad = c(26, 38, 56, 43),
    Nombre = c('Jorge', 'Paula', 'Loreto', 'Sebastián'),
    Tiene_mascota = c(FALSE, TRUE, TRUE, FALSE)
)
```

```r
df
```

```r
df$RUT <- c('523-5', '457-9', '185-K', '873-3')
```

```r
df
```

```r
names <- paste('f', 1:nrow(df), sep = '_')
names
```

```r
rownames(df) <- names
```

```r
df
```

```r
df['genero'] <- c('M', 'F', 'F', 'M')
df['Género'] <- c('M', 'F', 'F', 'M')
```

```r
df
```

```r
df <- df[, -5]
```

```r
df
```

```r
fila_nueva <- c(
    Edad = 34,
    Nombre = 'Pedro',
    Tiene_mascota = F,
    RUT = '125-0',
    Género = 'M')

rbind(df, fila_nueva)
```

```r
# Queremos identificar las personas que tienen mascota y tiene más de 35 años.

subconjunto_1 <- df[(df$Edad > 35) & (Tiene_mascota == TRUE),]
print(subconjunto_1)
```

```r
# Si queremos de la misma consulta anterior, pero sólo nombre y RUT.

subconjunto_2 <- df[(df$Edad > 35) & (Tiene_mascota == TRUE), c(2, 4)]
subconjunto_2

subconjunto_3 <- df[(df$Edad > 35) & (Tiene_mascota == TRUE), c('Nombre', 'RUT')]
subconjunto_3
```

```r
# Si quisiéramos ver los vuelos del 31 de Diciembre (todos son del mismo año).
subset_1 <- vuelos[(vuelos$month == 12) & (vuelos$day == 31), ]
subset_1
```

```r
# Lo mismo, pero con dplyr

subset_2 <- filter(
    vuelos,
    month == 12 & day == 31)
subset_2
```

```r
install.packages('compare')
library(compare)
```

```r
compare(subset_1, subset_2)
```
